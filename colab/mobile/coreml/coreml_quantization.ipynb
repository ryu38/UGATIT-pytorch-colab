```python
import coremltools as ct
from coremltools.models.neural_network import quantization_utils as quatils
```

```python
model_float32 = ct.models.MLModel(model_path)
```

```python
model_8bit = quatils.quantize_weights(model_float32, nbits=8)
# model_16bit = quatils.quantize_weights(model_float32, nbits=16)
```

```python
quant_model_name = 'ugatit-8bit.mlmodel'
quant_model_path = f'{model_dirname}/{quant_model_name}'
```

```python
ct.utils.save_spec(model_8bit, quant_model_path)
```
